```python
# Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import io
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv("/content/tested.csv")
```

```python
display(df.head())
```

```python
# Display entire dataset (use cautiously if dataset is large)
display(df)
```

```python
# ‚úÖ Step 2: Dataset Columns
print("\n‚ùï Columns in Dataset:")
for column in df.columns:
    print(f'- {column}')
```

```python
# ‚úÖ Step 3: Dataset Info
print("\n‚ÑπÔ∏è Dataset Info:")
df.info()
```

```python
# ‚úÖ Step 4: Statistical Summary
print("\nüìà Statistical Summary:")
display(df.describe(include='all'))
```

```python
# ‚úÖ Step 5: Missing Values Check
print("\n‚ùì Missing Values Count per Column:")
print(df.isnull().sum())
```

```python
# ‚úÖ Step 6: Calculate missing values (count & percentage)
missing = df.isnull().sum().sort_values(ascending=False)
missing_percent = (missing / len(df) * 100).round(2)

# Create DataFrame with Missing Count & Percent
missing_df = pd.concat([missing, missing_percent], axis=1, keys=['Missing Count', 'Missing Percent'])

# Display styled missing values table
display(missing_df.style.background_gradient(cmap='Reds'))
```

```python
# ‚úÖ Step 7: Count of Sex (Male & Female)
if 'Sex' in df.columns:
    sex_counts = df['Sex'].value_counts()
    print("\nüë®‚Äçü¶∞üë© Sex Counts:")
    print(sex_counts)

    # Pie Chart for Sex Distribution
    plt.figure(figsize=(6, 6))
    plt.pie(
        sex_counts,
        labels=[f"{sex} ({count})" for sex, count in zip(sex_counts.index, sex_counts.values)],
        autopct='%1.1f%%',
        startangle=90,
        colors=['#66b3ff', '#ff9999'],
        wedgeprops={'edgecolor': 'black'}
    )
    plt.title('üîπ Distribution of Sex')
    plt.show()
```

```python
# ‚úÖ Step 8: Countplot for Sex
if 'Sex' in df.columns:
    sex_counts = df['Sex'].value_counts()
    plt.figure(figsize=(8, 5))
    sns.countplot(x='Sex', data=df, palette="Set2")
    plt.title('üìä Count Plot: Sex Distribution')
    plt.xlabel('Sex')
    plt.ylabel('Count')
    plt.show()
```

```python
# ‚úÖ Step 9: Distribution of all numerical features
df.hist(figsize=(10, 8), bins=15, edgecolor='black')
plt.suptitle("üìä Feature Distributions", fontsize=14)
plt.show()
```

```python
# ‚úÖ Step 10: Scatter Plot (HP vs VOL if available)
if 'Age' in df.columns and 'Pclass' in df.columns and 'Sex' in df.columns:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df, x='Age', y='Pclass', hue='Sex', palette="viridis")
    plt.title('‚öôÔ∏è Age vs Pclass')
    plt.xlabel('Age')
    plt.ylabel('Pclass')
    plt.grid(True)
    plt.show()
```

```python
# ‚úÖ Step 10: Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='Blues', fmt=".2f")
plt.title("üìå Correlation Heatmap")
plt.show()
```
